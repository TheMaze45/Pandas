## 1. Select the name and state of all stores located in either California (CA) or Oregon (OR)

````python
# Lets have a look at the store table
stores
````

````python
# with this function we select all stores where either state == "CA" OR
# state == "OR" with the | (OR operator)
stores.loc[(stores["state"]=="CA") | (stores["state"]=="OR")]
````

````python
# As an alternative, we can use the .isin() function
stores.loc[stores['state'].isin(['CA', 'OR'])]
````

### Hint:

In SQL the syntax is:

```sql
SELECT stor_name,stor_address,state  
FROM stores  
WHERE state IN("CA","OR");
```

## 2. Using `.isin()`, select all titles of type "psychology", "mod_cook" and "trad_cook"

````python
#Lets have a look at the titles table
titles
````

````python
# Here we need the newly learned .isin() function
# Lets try it
# The basics are working, we get back a boolean return 
titles.isin(["psychology","mod_cook","trad_cook"])
````

````python
# We improved the function, now we get back all titles where on of the specified types is matched
titles.loc[titles["type"].isin(["psychology","mod_cook","trad_cook"])]
````

````python
# We can further clean it, to only display title and notes for example
titles[["title","notes"]].loc[titles["type"].isin(["psychology","mod_cook","trad_cook"])]
````

### Hint:

In SQL the syntax is:

```sql
SELECT title,type
FROM titles
WHERE type IN('psychology','mod_cook','trad_cook');
```

## 3. Select all the authors from the authors table that do not come from the cities Salt Lake City, Ann Arbor, and Oakland.

````python
# Lets have a look at the authors table
authors
````

````python
# Here we also need the .isin function, but we need to negate it with ~(TILDE sign)
# Normally the .isin gives us a return value where only the specified values are shown
authors.loc[authors["city"].isin(["Salt Lake City","Ann Arbor","Oakland"])]
````

````python
# Now the negation
# We see, that we get back every value BUT the ones that we specified
authors.loc[~authors["city"].isin(["Salt Lake City","Ann Arbor","Oakland"])]
````

### Hint:

In SQL the syntax is:

```sql
SELECT *
FROM authors
WHERE city NOT IN ('Salt Lake City', 'Ann Arbor', 'Oakland');
```

## 4. Select all the order numbers with a quantity sold between 25 and 45 from the table sales

````python
# Open up the sales table
sales
````

````python
# The .between() function offers everything we need
sales.loc[sales["qty"].between(25, 45)]
````

### Hint:

In SQL the syntax is:

```sql
SELECT ord_num, qty
FROM sales
WHERE qty BETWEEN 25 AND 45;
```

## 5. Select all the sales between 1993-03-11 and 1994-09-13

````python
# Open up the sales table again for better visibility
sales
````

````python
# Now we need to filter everything between two dates, pay attention to the format !
sales.loc[sales["ord_date"].between("1993-03-11", "1994-09-13",inclusive='both')]
````

````python
# As always, there are many ways that lead to Rome.
# Alternative solution 1:
sales.loc[(sales['ord_date'] >= '1993-03-11 00:00:00') & (sales['ord_date'] <= '1994-09-13 00:00:00')]
````

````python
Alternative solution 2:
````

### Hint:

In SQL the syntax is:

```sql
SELECT * 
FROM sales
WHERE ord_date BETWEEN "1993-03-11" AND "1994-09-13";
```

## 6. Select all job descriptions with a maximum level ("max_lvl") between 150 and 200.

````python
# Open up the jobs table
jobs
````

````python
# Again the .loc() function
# We see that nobody falls into this specific category
# But the function works, if we raise the upper limit
jobs.loc[(jobs["min_lvl"] >= 150) & (jobs["max_lvl"] <= 200)]
````

````python
# Now we have some values
jobs.loc[(jobs["min_lvl"] >= 150) & (jobs["max_lvl"] <= 300)]
````

````python
# Looking at the solutions, I made a filtering mistake, the question specified a MAXIMUM LEVEL between 150 and 200
# That changes everything a bit, here the .between() function would be best
jobs.loc[jobs['max_lvl'].between(150, 200)]
````

### Hint:

In SQL the syntax is:

```sql

FROM jobs
```
