## 1. Select all books from the table `titles` that contain the word "cooking" in its title

## 3. Select the full names (first and last name) of authors whose last name starts with "S"

````python
# Lets see the authors table
# We see 3 authors whose last names starts with an "S"
authors
````

````python
# Here we selected every author whose last name starts with an "S" and only display the columns "au_fname" & "au_lname"
authors[['au_fname',"au_lname"]].loc[(authors["au_lname"].str.startswith("S"))]
````

## 4. Select the name and address of all stores located in an Avenue (its address ends with "Ave.")

````python
# Lets have a look at the stores table
stores
````

````python
# We can also use the .loc function here, we only need two change on or two things
# Here we selected to display the store name and store adress columns of all rows where the string of the store adress ends with "Ave."
stores[['stor_name',"stor_address"]].loc[(stores["stor_address"].str.endswith("Ave."))]
````

### Hint:

In SQL the syntax is:

```sql
SELECT stor_name, stor_address
FROM stores
WHERE stor_address LIKE ("%Ave.");
```

## 5. Select the name and address of all stores located in an Avenue or in a Street (address ended in "St.")

````python
# Here we need more than one condition, we want to select every row where either the condition ends with "Ave." or "St." is True.
# We can connect conditions with | (OR operator) and also & (AND operator)
stores[['stor_name',"stor_address"]].loc[(stores["stor_address"].str.endswith("Ave.")) | (stores["stor_address"].str.endswith("St."))]
````

### Hint:

In SQL the syntax is:

```sql
SELECT
	stor_name,
	stor_address
FROM
	stores
WHERE 
	stor_address LIKE '%St.' 
    OR stor_address LIKE '%Ave.';
```

## 6. Look at the "employee" table (select all columns to explore the raw data):
   Find a pattern that reveals whether an employee is Female or Male.
   Select all female employees.

````python
# Lets have a look at the employee table
employee
````

````python
# We can see that every female employee has an capital "F" at the very end in their "emp_id"
# So we can filter with that criteria
employee.loc[employee["emp_id"].str.endswith("F")]
````

### Hint:

In SQL the syntax is:

```sql
SELECT emp_id, fname,lname 
FROM employee 
WHERE emp_id LIKE ('%F');
```

## 7. Select the first and last names of all male employees whose name starts with "P".

````python
# First let us select all male employees, just copy & paste the .loc from the exercise before and change "F" to "M"
employee.loc[employee["emp_id"].str.endswith("M")]
````

````python
# Now let us further refine the filter and only show the data we want
# Here again multiple conditions that we want to filter, but not as before with the | (OR operator)
# for this case we need the & (AND operator)
employee[["fname","lname"]].loc[(employee["emp_id"].str.endswith("M")) & (employee["fname"].str.startswith("P"))]
````

### Hint:

In SQL the syntax is:

```sql
SELECT
    emp_id,
    fname,
    lname
FROM
    employee
WHERE
    (fname LIKE 'P%' 
    OR lname LIKE 'P%') 
    AND emp_id LIKE '%M';
```
